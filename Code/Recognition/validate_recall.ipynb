测试 llama是否能将gpt识别为AI的也都识别为AI

```python
import json

tar_files = [
    '../data/extract_by_llm/llama_8B_v3/compare_gpt.json',
    '../data/extract_by_llm/Qwen_32B_AWQ_v3/compare_gpt.json',
    '../data/extract_by_llm/Qwen_32B_v3/compare_gpt.json',
    '../data/extract_by_llm/Qwen_72B_AWQ_v3/compare_gpt.json',
    '../data/extract_by_llm/llama_70B_AWQ_v3/compare_gpt.json',
]
with open(tar_files[3], 'r') as f:
    datas = json.load(f)

len(datas)
```

```python
cnt = 0
not_ai_dois = []
ai_dois = []

for id, data in datas.items():
    doi = data['doi']
    doi = doi.replace('https://doi.org/', '')
    if data['tag'] == 'Yes':
        cnt += 1
        ai_dois.append(doi)
    else:
        not_ai_dois.append(doi)

all_dois =  not_ai_dois + ai_dois

cnt
```

```python
with open("../data/from_old_project/ai_identify_v1_res.json", 'r') as f:
    v1_ai_datas = json.load(f)
    v1_ai_dois = []
    for doi, data in v1_ai_datas.items():
        if data['uses_artificial_intelligence_method'] == 'Yes':
             v1_ai_dois.append(doi)

with open("../data/from_old_project/ai_identify_v2_res.json", 'r') as f:
    v2_ai_datas = json.load(f)
    v2_ai_dois = []
    for doi, data in v2_ai_datas.items():
        if data['contain_artificial_intelligence_term'] == 'Yes':
             v2_ai_dois.append(doi)

with open("../data/from_old_project/ai_identify_v3_res.json", 'r') as f:
    v3_ai_datas = json.load(f)
    v3_ai_dois = []
    for doi, data in v3_ai_datas.items():
        if data['contain_artificial_intelligence_term'] == 'Yes':
             v3_ai_dois.append(doi)

with open("../data/from_old_project/ai_identify_v4_res.json", 'r') as f:
    v4_ai_datas = json.load(f)
    v4_ai_dois = []
    for doi, data in v4_ai_datas.items():
        if data['contain_artificial_intelligence_term'] == 'Yes':
             v4_ai_dois.append(doi)
```

```python
print(len(set(v1_ai_dois)), len(set(v1_ai_dois) & set(all_dois)), len(set(v1_ai_dois) & set(ai_dois)))
print(len(set(v2_ai_dois)), len(set(v2_ai_dois) & set(all_dois)), len(set(v2_ai_dois) & set(ai_dois)))
print(len(set(v3_ai_dois)), len(set(v3_ai_dois) & set(all_dois)), len(set(v3_ai_dois) & set(ai_dois)))
print(len(set(v4_ai_dois)), len(set(v4_ai_dois) & set(all_dois)), len(set(v4_ai_dois) & set(ai_dois)))
```
